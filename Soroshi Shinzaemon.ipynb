Problem 1


```python
grain = 1
grain_per_day = []
for day in range(100):
  if day == 0:
    grain_per_day.append(grain)
  else:
    grain_per_day.append((grain*2)**day)
print("the total rice grains after 100 days: ", sum(grain_per_day))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.title("Grains per day")
plt.xlabel("days")
plt.ylabel("grains")
plt.plot(grain_per_day)
plt.show()
```

Problem 2

```python
# A function that calculates the number of rice grains per day in the rice anecdote of Shinzaemon Sorori.
def compute_sorori_shinzaemon(days = 100):
  grain = 1
  grain_per_day2 = []
  for day in range(days):
    if day == 0:
      grain_per_day2.append(grain)
    else:
      grain_per_day2.append((grain*2)**day)
  print("the total rice grains after days: ", sum(grain_per_day2))

  # Graph visualization for Problem 2
  plt.title("Grains per day")
  plt.xlabel("days")
  plt.ylabel("grains")
  plt.plot(grain_per_day2)
  plt.show()
```

```python
compute_sorori_shinzaemon(80)
```
